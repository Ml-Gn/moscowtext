```python
import spacy, csv
from spacy.matcher import Matcher

nlp = spacy.load("ru_core_news_sm")
nlp.max_length = 20000000
matcher = Matcher(nlp.vocab)


pattern = [{"ENT_TYPE": "LOC"}]
matcher.add("Location", [pattern])
pattern = [{"LEMMA": "улица"}, {"IS_TITLE": True}]
matcher.add("Улица какая-то", [pattern])
pattern = [{"IS_TITLE": True}, {"LEMMA": "улица"}]
matcher.add("Какая-то улица", [pattern])
pattern = [{"POS": "ADJ"}, {"LEMMA": "проспект"}]
matcher.add("какой-то проспект", [pattern])
pattern = [{"LEMMA": "проспект"}, {"POS": "ADJ"}]
matcher.add("Проспект какой-то", [pattern])
pattern = [{"IS_TITLE": True}, {"LEMMA": "бульвар"}]
matcher.add("Какой то бульвар", [pattern])
pattern = [{"LEMMA": "бульвар"}, {"IS_TITLE": True}]
matcher.add("Бульвар какой-то", [pattern])
pattern = [{"IS_TITLE": True}, {"LEMMA": "Аллея"}]
matcher.add("Какая то аллея", [pattern])
pattern = [{"LEMMA": "Аллея"}, {"IS_TITLE": True}]
matcher.add("Аллея какая то", [pattern])
pattern = [{"IS_TITLE": True}, {"LEMMA": "переулок"}]
matcher.add("Какой то переулок", [pattern])
pattern = [{"LEMMA": "переулок"}, {"IS_TITLE": True}]
matcher.add("Переулок какой-то", [pattern])
pattern = [{"IS_TITLE": True}, {"LEMMA": "набережная"}]
matcher.add("Какой то набережная ", [pattern])
pattern = [{"LEMMA": "набережная"}, {"IS_TITLE": True}]
matcher.add(" какой-то набережная", [pattern])
pattern = [{"IS_TITLE": True}, {"LEMMA": "проезд"}]
matcher.add("Какой то проезд", [pattern])
pattern = [{"LEMMA": "проезд"}, {"IS_TITLE": True}]
matcher.add(" проезд какой-то ", [pattern])
pattern = [{"IS_TITLE": True}, {"LEMMA": "тупик"}]
matcher.add("Какой то ", [pattern])
pattern = [{"LEMMA": "тупик"}, {"IS_TITLE": True}]
matcher.add("тупик какой-то", [pattern])
pattern = [{"IS_TITLE": True}, {"LEMMA": "шоссе"}]
matcher.add("Какой то шоссе", [pattern])
pattern = [{"LEMMA": "бульвар"}, {"IS_TITLE": True}]
matcher.add("бульвар какой-то", [pattern])
pattern = [{"IS_TITLE": True}, {"LEMMA": "бульвар"}]
matcher.add("какой-то бульвар", [pattern])
# наши правила, паттерны поиска
# в фигурных скобках атрибуты правила
# IS_TITLE - означает что первая буква должна быть заглавной
# LEMMA - означает любой вариант слова в кавычках
# matcher.add("Это название правила", [pattern]) - тут мы добавляем паттерн в общий список правил
# полный список атрибутов через которые могут работать правил можно найти по ссылке https://spacy.io/api/matcher

file_name = 'glavy.txt'
introduction_file_text = open(file_name).read()
introduction_file_doc = nlp(introduction_file_text)
# открываем файл

header = ['named entity', 'start', 'stop']
rows = []
# заготовка для csv: заголовок и пустой список
for ent in introduction_file_doc.ents:
    if ent.label_ == 'LOC':
        one_row = []
        one_row = [ent.lemma_, ent.start_char, ent.end_char]
        rows.append(one_row)
# цикл в котором мы идем по определившимся сущностям
# если сущность - локация, записываем лемматизированную версию в one_row
# потом собираем все списки one_row в один большой список со списками - rows
```

```python
rows
# смотрим на результат
```

```python
with open('results_spacy.csv','w') as file:
  writer = csv.writer(file)
  writer.writerow(header)
  writer.writerows(rows)
# записываем в отдельный файл csv наш огромный список row через библиотеку сsv
# файл появится через ~минуту слева в files в той папке где мы работаем
```

```python
file_name = 'glavy.txt'
introduction_file_text = open(file_name).read()
introduction_file_doc = nlp(introduction_file_text)
matches = matcher(introduction_file_doc)
# открываем файл
# _text и _doc это внутренние приколы спейси, по ним определяется может ли спейси с ними работать
# https://spacy.io/api/doc
header = ['named entity', 'start', 'stop']
rows = []
# заготовка для перевода в csv
for match_id, start, end in matches:
    first_row = []
    string_id = nlp.vocab.strings[match_id]  
    span = introduction_file_doc[start:end]
    first_row = [span.lemma_, start, end]  
    rows.append(first_row)
# проходимся по файлу через заданные паттерны правил (ячейки выше)
# циклично записываем совпадения в списки 'first_row'
# засовываем все списки в один большой 'row'
```

```python
rows
# Смотрим что получилось
```

```python
with open('results_spacy_rules.csv','w') as file:
  writer = csv.writer(file)
  writer.writerow(header)
  writer.writerows(rows)
# через библиотеку csv записываем получившийся список row в csv формат в отдельный файл
```
